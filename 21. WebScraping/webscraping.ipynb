# WebScraping

```python
# importing required libraries
import requests
from bs4 import BeautifulSoup
```

```python
r = requests.get("http://www.pyclass.com/example.html", 
                 headers={'User-agent': 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:61.0) Gecko/20100101 Firefox/61.0'})
c=r.content
```

```python
c
```

```python
# for displaying the extracted data in html format
soup=BeautifulSoup(c,"html.parser")
print(soup.prettify())
```

```python
# Extracting cities
all=soup.find_all("div",{"class":"cities"})
h=all
all
```

```python
type(all)
```

```python
# Extracting only "London"
all=soup.find_all("div",{"class":"cities"})[0]
print(all.prettify())
```

```python
all.find_all("h2")
```

```python
all.find_all("h2")[0]
```

```python
# Extracting text data 
all.find_all("h2")[0].text
```
